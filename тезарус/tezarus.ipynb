## Блок 1 - Введение в pandas <a id="блок1"></a>

---

### Основные структуры

<a id="series"></a>
**Series** - одномерный массив с индексом. Как один столбец таблицы. Создаётся через `pd.Series([...], index=[...])`. [docs](https://pandas.pydata.org/docs/reference/api/pandas.Series.html)

<a id="dataframe"></a>
**DataFrame** - двумерная таблица: строки и столбцы, у каждого столбца своё имя. Основная структура в pandas. Каждый столбец - это Series. [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html)


<a id="атрибут"></a>
**Атрибут** - свойство объекта, вызывается через точку без скобок. Примеры: `df.shape`, `df.columns`, `df.dtypes`, `series.values`, `series.index`.

<a id="метод"></a>
**Метод** - функция объекта, вызывается через точку со скобками. Примеры: `df.head()`, `df.info()`, `df.describe()`.

---

### Форматы входных данных

<a id="csv"></a>
**CSV** (Comma-Separated Values) - текстовый формат: значения разделены запятой. Самый распространённый формат для передачи табличных данных. `pd.read_csv()`

<a id="excel"></a>
**Excel (.xlsx)** - табличный формат Microsoft. `pd.read_excel()`

<a id="json"></a>
**JSON** - формат для вложенных структур. `pd.read_json()` [docs](https://pandas.pydata.org/docs/user_guide/io.html)

---

### Первичный осмотр датасета

<a id="head"></a>
**`.head(n)`** - первые `n` строк (по умолчанию 5). `.tail(n)` - последние `n`.

<a id="info"></a>
**`.info()`** - типы столбцов, количество непустых значений, объём памяти. Первый метод при знакомстве с данными.

<a id="describe"></a>
**`.describe()`** - описательная статистика по числовым столбцам: count, mean, std, min, max, квантили. Для категориальных - `df.describe(include='object')`.

<a id="shape"></a>
**`.shape`** - кортеж `(строки, столбцы)`. Атрибут, не метод.

<a id="columns"></a>
**`.columns`** - список названий столбцов.

<a id="dtypes"></a>
**`.dtypes`** - типы данных каждого столбца (int64, float64, object и т.д.).

---

### Распределение значений

<a id="value_counts"></a>
**`.value_counts()`** - частоты каждого уникального значения в столбце, по убыванию. Удобно для категориальных данных.

<a id="unique"></a>
**`.unique()`** - массив уникальных значений столбца.

<a id="nunique"></a>
**`.nunique()`** - количество уникальных значений (число).

---

### Фильтрация

<a id="логическая-индексация"></a>
**Логическая индексация** - фильтрация строк по условию: `df[df['col'] > 30]`. Условие возвращает булев Series (True/False), DataFrame оставляет только строки с True.

<a id="loc"></a>
**`.loc[]`** - выбор строк и столбцов по меткам: `df.loc[условие, ['col1', 'col2']]`. Условия комбинируются через `&` и `|`, каждое в скобках.

---

### Создание новых столбцов

<a id="присваивание-столбца"></a>
**Прямое присваивание** - `df['new_col'] = значения`. Изменяет исходный DataFrame. Самый частый способ.

<a id="apply"></a>
**`.apply(func)`** - применяет функцию к каждому элементу столбца или к каждой строке. Используется когда нужна своя логика: `df['col'].apply(lambda x: ...)`.

<a id="assign"></a>
**`.assign()`** - создаёт новый столбец без изменения оригинала, возвращает новый DataFrame: `df.assign(new_col=df['a'] + df['b'])`.

---

### Агрегация

<a id="groupby"></a>
**`.groupby(col)`** - группировка строк по значениям столбца. После группировки применяется агрегирующая функция: `df.groupby('sex')['age'].mean()`.

<a id="agg"></a>
**`.agg()`** - несколько агрегаций сразу: `df.groupby('col').agg({'age': 'mean', 'fare': 'sum'})`.

Основные агрегирующие функции: `.mean()`, `.median()`, `.sum()`, `.count()`, `.min()`, `.max()`, `.std()`.

## Блок 2 - Типы проблем в данных и диагностика <a id="блок2"></a>

---

### Типы проблем

<a id="пропуск"></a>
**Пропуск (NaN, missing value)** - отсутствующее значение в ячейке. Обозначается как `NaN` (Not a Number) в pandas. Возникает из-за ошибок сбора данных, незаполненных полей. Смещает средние, уменьшает выборку.

<a id="дубликат"></a>
**Дубликат** - полный или частичный повтор строки. Завышает частоты и искажает статистику.

<a id="аномалия"></a>
**Аномалия (выброс)** - значение, которое сильно отклоняется от остальных. Бывает трёх видов:
- ошибка измерения/ввода (отрицательная цена) - удалить
- редкое, но реальное наблюдение (очень дорогое вино) - сохранить или обработать отдельно
- систематическое искажение - исследовать источник

<a id="несогласованный-текст"></a>
**Несогласованный текст** - одно и то же значение записано по-разному: разный регистр, лишние пробелы, опечатки, синонимы. Мешает корректной группировке (`'US'`, `'USA'`, `'U.S.A.'` - одно и то же).

---

### Диагностика

<a id="isna"></a>
**`.isna()`** / **`.isnull()`** - булева маска: True там где NaN. `.isna().sum()` - количество пропусков по каждому столбцу. [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isna.html)

<a id="duplicated"></a>
**`.duplicated()`** - булева маска: True для строк-дубликатов. `.duplicated().sum()` - общее число дубликатов. [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.duplicated.html)

<a id="drop"></a>
**`.drop(columns=[...])`** - удаление столбцов или строк по меткам. [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html)

## Блок 3 - Очистка данных <a id="блок3"></a>

---

### Дубликаты

<a id="drop_duplicates"></a>
**` .drop_duplicates()`** - удаляет повторяющиеся строки. По умолчанию смотрит на все столбцы, но можно указать подмножество: `df.drop_duplicates(subset=['col'])` . [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html)

---

### Пропуски

<a id="dropna"></a>
**`.dropna()`** - удаляет строки с пропусками. Радикальный способ - теряем данные. [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html)

<a id="fillna"></a>
**`.fillna(value)`** - заполняет пропуски значением: константой, средним, медианой, строкой. Пример: `df['col'].fillna(df['col'].median())` . [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html)

<a id="transform"></a>
**`.transform(func)`** - применяет функцию внутри каждой группы, возвращает Series того же размера. Удобно для заполнения пропусков средним по группе: `df.groupby('col')['price'].transform('mean')` . [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.transform.html)

---

### Аномалии

<a id="iqr"></a>
**IQR (межквартильный размах)** - разница между 75-м и 25-м перцентилями: IQR = Q3 - Q1. Выбросы - значения ниже Q1 - 1.5*IQR или выше Q3 + 1.5*IQR.

<a id="zscore"></a>
**Z-score** - сколько стандартных отклонений значение отстоит от среднего: z = (x - mean) / std. Значения с |z| > 3 обычно считаются выбросами. `.scipy.stats.zscore()` [docs](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.zscore.html)

<a id="quantile"></a>
**`.quantile(q)`** - перцентиль уровня q (от 0 до 1). Например `df['price'].quantile(0.95)` - 95-й перцентиль. Используется для задания порогов при фильтрации выбросов. [docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.quantile.html)

---

### Очистка текста

<a id="str-strip"></a>
**`.str.strip()`** - убирает пробелы по краям строки. [docs](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.strip.html)

<a id="str-lower"></a>
**`.str.lower()`** - переводит все символы в нижний регистр. `.str.upper()` - в верхний. [docs](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.lower.html)

<a id="str-replace"></a>
**`.str.replace(pat, repl)`** - замена подстроки или паттерна. Поддерживает регулярные выражения. [docs](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.replace.html)

<a id="str-contains"></a>
**`.str.contains(pat)`** - булева маска: True если строка содержит паттерн. Удобно для фильтрации по тексту. [docs](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.contains.html)

## Блок 4 - Разведочный анализ (EDA) <a id="блок4"></a>

---

<a id="eda"></a>
**EDA (Exploratory Data Analysis, разведочный анализ)** - первый этап работы с данными: знакомство с распределениями, поиск аномалий, изучение связей между признаками. Цель - понять данные до построения моделей или выводов.

---

### Одномерный анализ

Анализ каждого признака по отдельности. Опирается на меры центральности, меры разброса и форму распределения.

**Меры центральности**

<a id="среднее"></a>
**Среднее** (.mean()) - сумма значений делённая на количество. Чувствительно к выбросам.

<a id="медиана"></a>
**Медиана** (.median()) - значение, которое делит отсортированный ряд пополам. Устойчива к выбросам - предпочтительна при асимметричных распределениях.

<a id="мода"></a>
**Мода** (.mode()) - наиболее часто встречающееся значение. Используется для категориальных данных или мультимодальных распределений.

---

**Меры разброса**

<a id="дисперсия"></a>
**Дисперсия** (.var()) - среднее квадратичное отклонение от среднего. [docs](https://pandas.pydata.org/docs/reference/api/pandas.Series.var.html)

<a id="стандартное-отклонение"></a>
**Стандартное отклонение** (.std()) - корень из дисперсии. В тех же единицах что и данные, поэтому интерпретировать проще. [docs](https://pandas.pydata.org/docs/reference/api/pandas.Series.std.html)

---

**Форма распределения**

<a id="асимметрия"></a>
**Асимметрия** (.skew()) - мера скошенности распределения. Положительная (>0) - длинный правый хвост, среднее > медианы. Отрицательная - длинный левый хвост. При |skew| > 1 лучше использовать медиану вместо среднего. [docs](https://pandas.pydata.org/docs/reference/api/pandas.Series.skew.html)

<a id="эксцесс"></a>

### Разделение данных

<a id="train-test-split"></a>
**train_test_split()** - делит датасет на обучающую и тестовую выборки. Параметр random_state фиксирует сид для воспроизводимости.

<a id="утечка-данных"></a>
**Утечка данных (data leakage)** - когда информация из теста попадает в обучение. Правило: fit() вызывается только на train, transform() - и на train, и на test с теми же параметрами. Если сделать fit() на всём датасете - модель "знает" статистики теста, результаты будут завышены.

<a id="целевая-переменная"></a>
**Целевая переменная (target, y)** - то, что предсказываем. **Признаки (features, X)** - то, по чему предсказываем.

### Кодирование категориальных признаков

<a id="one-hot-encoding"></a>
**One-Hot Encoding** - создаёт отдельный бинарный столбец для каждой категории (1 если равно, 0 иначе). Не создаёт ложного порядка. Увеличивает размерность. Подходит для номинальных признаков (цвет, страна). [docs](https://www.geeksforgeeks.org/machine-learning/one-hot-encoding-vs-label-encoding/)

<a id="ordinal-encoding"></a>
**Ordinal (Label) Encoding** - присваивает каждой категории число (0, 1, 2...). Создаёт порядок между категориями - опасно для линейных моделей если порядка нет. Подходит для ординальных признаков (S < M < L) и деревьев.

<a id="номинальный"></a>
**Номинальный признак** - категории без порядка (цвет, страна). Нужен One-Hot.

<a id="ординальный"></a>
**Ординальный признак** - категории с естественным порядком (огранка: Fair < Good < Ideal). Можно Ordinal Encoding.

### Нормализация и масштабирование

<a id="standardscaler"></a>
**StandardScaler** - стандартизация: вычитает среднее, делит на std. Результат: среднее=0, std=1. Обязателен для линейных моделей и KMeans. [docs](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)

<a id="пайплайн"></a>
**Pipeline** - цепочка трансформаций + модель. Упрощает код, исключает утечку данных, позволяет применять разную обработку к разным столбцам (ColumnTransformer).

### Метрики регрессии

<a id="mae"></a>
**MAE (Mean Absolute Error)** - средняя абсолютная ошибка:

$$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

Показывает на сколько единиц в среднем ошибается модель. Устойчива к выбросам - большие ошибки не доминируют.

<a id="rmse"></a>
**RMSE (Root Mean Squared Error)** - корень из средней квадратичной ошибки:

$$RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

Сильнее штрафует за большие ошибки чем MAE из-за возведения в квадрат.

<a id="r2"></a>
**R² (коэффициент детерминации)** - доля дисперсии целевой переменной, объяснённая моделью:

$$R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}$$

1.0 = идеально, 0 = не лучше простого среднего, <0 = хуже среднего.


---

### Метрики классификации

<a id="confusion-matrix"></a>
**Confusion Matrix (матрица ошибок)** - таблица 2x2:

| | Предсказано 0 | Предсказано 1 |
|---|---|---|
| **Факт 0** | TN (верно отриц.) | FP (ложно полож.) |
| **Факт 1** | FN (ложно отриц.) | TP (верно полож.) |

<a id="accuracy"></a>
**Accuracy** - доля правильных ответов из всех:

$$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$

Бесполезна при дисбалансе. Если 73% клиентов не уходят, модель которая всегда говорит "не уйдёт" получит accuracy 73% - но не найдёт ни одного реального ухода.

<a id="precision"></a>
**Precision** - из всех предсказанных положительных, сколько реально положительных:

$$Precision = \frac{TP}{TP + FP}$$

Важна когда дорого ошибочно обвинить.

<a id="recall"></a>
**Recall** - из всех реально положительных, сколько нашла модель:

$$Recall = \frac{TP}{TP + FN}$$

Важна когда дорого пропустить (болезнь у пациента).

<a id="f1"></a>
**F1-score** - гармоническое среднее precision и recall:

$$F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}$$

Баланс между precision и recall. Используется при дисбалансе вместо accuracy.

<a id="дисбаланс-классов"></a>
**Дисбаланс классов** - когда одного класса значительно больше. Смотреть на F1, precision, recall - не на accuracy.


---

### Кластеризация

<a id="центроид"></a>
**Центроид** - центр кластера. Точка в пространстве признаков, координаты которой = среднее всех точек кластера. После каждой итерации пересчитывается заново.

<a id="kmeans"></a>
**KMeans** - алгоритм кластеризации. Работает итеративно:
1. Случайно инициализирует k центроидов
2. Каждую точку назначает ближайшему центроиду (по евклидову расстоянию)
3. Пересчитывает центроиды как среднее точек своего кластера
4. Повторяет шаги 2-3 до сходимости (центроиды перестают двигаться)

Нормализация обязательна - иначе признак с большим диапазоном (доход в тысячах) полностью доминирует над признаком с маленьким (возраст), и KMeans "видит" только его.

<a id="inertia"></a>
**Inertia** - сумма квадратов расстояний от каждой точки до центроида её кластера:

$$Inertia = \sum_{k} \sum_{x \in C_k} ||x - \mu_k||^2$$

Чем меньше - тем плотнее кластеры. При увеличении k всегда уменьшается. Используется для выбора оптимального k методом локтя: ищем точку где убывание резко замедляется.

<a id="метод-локтя"></a>
**Метод локтя** - строим график inertia от k. Оптимальное k - там где кривая "перегибается" как локоть: дальнейшее увеличение k почти не снижает inertia.

<a id="silhouette"></a>
**Silhouette score** - мера качества кластеризации для каждой точки:

$$s = \frac{b - a}{\max(a, b)}$$

где a = среднее расстояние до точек своего кластера, b = среднее расстояние до точек ближайшего чужого кластера. Итоговый score = среднее по всем точкам. Диапазон от -1 до 1: близко к 1 - точка хорошо вписывается в свой кластер, близко к 0 - на границе двух кластеров, отрицательное - скорее всего в неправильном кластере. [docs](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.silhouette_score.html)

### Разбиение данных

<a id="train-valid-test"></a>
**Train / Validation / Test** - три выборки:
- **Train** - на ней обучается модель
- **Validation** - для подбора гиперпараметров и выбора модели в процессе обучения
- **Test** - финальная оценка, трогается только один раз в самом конце

Validation нужен чтобы не "подглядывать" в тест при подборе параметров. Если настраивать по тесту - получим завышенные оценки.

<a id="stratify"></a>
**Stratify** - параметр в train_test_split. Сохраняет пропорции классов целевой переменной в каждой выборке. Работает только с категориальными признаками. [docs](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

<a id="kfold"></a>
**K-Fold кросс-валидация** - данные делятся на k частей (fold). Модель обучается k раз: каждый раз одна часть - валидация, остальные - train. Итоговая метрика = среднее по k запускам. Надёжнее одного разбиения. [docs](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html)

<a id="stratified-kfold"></a>
**StratifiedKFold** - K-Fold с сохранением пропорций классов в каждом fold. Используется при дисбалансе классов. [docs](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedKFold.html)

### Статистическое тестирование признаков

<a id="p-value"></a>
**p-value** - вероятность получить наблюдаемый результат (или более экстремальный), если нулевая гипотеза верна.

- **Нулевая гипотеза (H₀)** - утверждение об отсутствии эффекта/различия ("распределения одинаковые", "средние равны")
- **p-value < 0.05** - результат статистически значим, отвергаем H₀
- **p-value > 0.05** - нет оснований отвергать H₀ (но это не значит что H₀ верна)

Важно: p-value говорит только "есть ли различие?", но не "насколько оно велико?" - для этого нужен размер эффекта.

<a id="ks-test"></a>
**KS-тест (Kolmogorov-Smirnov)** - проверяет совпадают ли два распределения. Используется для проверки корректности разбиения train/test. p-value > 0.05 = нет оснований считать распределения разными. [docs](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ks_2samp.html)

<a id="ttest"></a>
**t-тест** - проверяет равенство средних двух групп. p-value > 0.05 = средние статистически не различаются. [docs](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html)

<a id="размер-эффекта"></a>
**Размер эффекта (effect size)** - числовая мера практической значимости различия между группами. В отличие от p-value, не зависит от размера выборки. p-value отвечает на вопрос "есть ли различие?", размер эффекта - "насколько оно велико?". При большой выборке даже ничтожная разница даёт p < 0.05, поэтому p-value без размера эффекта недостаточно.

<a id="cohens-d"></a>
**Cohen's d** - стандартизированный размер эффекта: разница средних двух групп, выраженная в единицах объединённого стандартного отклонения:

$$d = \frac{\mu_1 - \mu_2}{\sigma_{pooled}}$$

где $\mu_1$, $\mu_2$ - средние двух групп, $\sigma_{pooled}$ - объединённое стандартное отклонение (среднее std обеих групп).

Интерпретация: |d| < 0.2 = малый эффект, 0.2-0.5 = средний, > 0.8 = большой. Не зависит от размера выборки в отличие от p-value.

<a id="хи-квадрат"></a>
**Хи-квадрат ($\chi^2$)** - статистика для проверки связи между двумя категориальными переменными.

Идея: строим **таблицу сопряжённости** - сколько раз встречается каждая комбинация категорий. Например, smoker (yes/no) × region (north/south/east/west).

Если переменные **независимы**, то ожидаемое количество в каждой ячейке = (сумма строки × сумма столбца) / общее число. Хи-квадрат считает насколько реальные числа отличаются от ожидаемых:

$$\chi^2 = \sum \frac{(O - E)^2}{E}$$

где $O$ - реальное (observed) значение в ячейке, $E$ - ожидаемое (expected). Чем больше отклонение от ожидаемого - тем больше $\chi^2$ - тем сильнее связь. `scipy.stats.chi2_contingency()`.

**Проблема**: $\chi^2$ растёт с размером выборки. На 10 000 наблюдениях слабая связь даст большой $\chi^2$, на 100 - маленький. Сравнивать между разными датасетами нельзя.

<a id="cramers-v"></a>
**Cramer's V** - нормализованная версия $\chi^2$, от 0 до 1:

$$V = \sqrt{\frac{\chi^2}{n \cdot (k - 1)}}$$

Делим на $n$ (убираем влияние размера выборки) и на $(k-1)$ (убираем влияние числа категорий, $k$ = минимум из числа строк и столбцов). Получаем число в [0, 1]: 0 = связи нет, 1 = зная один признак, точно знаем другой. [wiki](https://en.wikipedia.org/wiki/Cram%C3%A9r%27s_V)

### Feature Engineering

<a id="feature-engineering"></a>
**Feature Engineering (конструирование признаков)** - создание новых признаков из существующих для улучшения качества модели.

<a id="полиномиальные-признаки"></a>
**Полиномиальные признаки** - возведение в степень: age², bmi². Улавливают нелинейные зависимости.

<a id="признаки-взаимодействия"></a>
**Признаки взаимодействия (interaction features)** - произведение двух признаков: age * smoker. Улавливают совместное влияние. Например: курение с возрастом влияет сильнее, чем курение само по себе.

<a id="логарифмическое-преобразование"></a>
**Логарифмическое преобразование** - log(x). Уменьшает влияние выбросов, сжимает большие значения, помогает при правосторонней асимметрии.

<a id="бинарный-признак"></a>
**Бинарный признак** - 0 или 1. Например: is_obese = 1 если bmi >= 30. Позволяет выделить подгруппу с особым поведением.

---

### Важность признаков через модели

<a id="feature-importance"></a>
**Feature Importance** - оценка вклада каждого признака в предсказание модели. Зависит от типа модели:
- **Линейные модели** (LinearRegression, Lasso) - важность = абсолютное значение коэффициента
- **Деревья** (RandomForest, CatBoost) - важность = насколько признак уменьшает ошибку при разбиениях

<a id="регуляризация"></a>
**Регуляризация** - добавление штрафа за большие коэффициенты к функции потерь. Не даёт модели "зацепиться" за шум в данных.

Обычная линейная регрессия минимизирует только ошибку:

$$\sum(y_i - \hat{y}_i)^2$$

<a id="lasso"></a>
**Lasso (L1-регуляризация)** - добавляет штраф за сумму абсолютных значений коэффициентов:

$$\sum(y_i - \hat{y}_i)^2 + \alpha \sum|w_j|$$

где $w_j$ - коэффициенты модели, $\alpha$ - сила регуляризации. Модель вынуждена одновременно уменьшать ошибку и держать коэффициенты маленькими. Особенность L1: незначимые коэффициенты обнуляются точно в ноль - автоматический отбор признаков. Чем больше $\alpha$, тем больше признаков обнуляется.

<a id="проклятие-размерности"></a>

### PCA (Principal Component Analysis)

<a id="pca"></a>
**PCA** - линейный метод снижения размерности. Находит новые оси (главные компоненты) в направлениях максимальной дисперсии данных и проецирует данные на них.

Формально: ищем матрицу $W \in \mathbb{R}^{D \times d}$, такую что проекция $Y = XW$ сохраняет максимум дисперсии. Столбцы $W$ - собственные векторы матрицы ковариации $X^TX$, отсортированные по убыванию собственных значений.

**Как применять:**
```python
pca = PCA(n_components=50)
X_train_pca = pca.fit_transform(X_train)  # fit только на train
X_test_pca = pca.transform(X_test)
```

<a id="главные-компоненты"></a>
**Главные компоненты** - новые признаки после PCA. Линейные комбинации исходных. Первая компонента объясняет наибольшую долю дисперсии, каждая следующая - меньше.

<a id="объяснённая-дисперсия"></a>
**Объяснённая дисперсия (explained variance ratio)** - доля исходной дисперсии, сохранённая каждой компонентой. `pca.explained_variance_ratio_`. Сумма по всем компонентам = сколько информации сохранено.

### t-SNE

<a id="tsne"></a>
**t-SNE** - нелинейный метод снижения размерности для визуализации. Сохраняет локальную структуру: близкие точки в исходном пространстве остаются близкими.

Алгоритм:
1. В исходном пространстве вычисляет вероятности сходства каждой пары точек (гауссово ядро)
2. В низкой размерности строит похожие вероятности (t-распределение Стьюдента)
3. Минимизирует KL-дивергенцию между двумя распределениями

**Особенности:**
- Только для визуализации (нет `transform` - нельзя применить к новым данным)
- Медленно работает на высокой размерности - сначала применять PCA
- Хорошо выявляет кластеры, но расстояния между кластерами не интерпретируются

### UMAP

<a id="umap"></a>
**UMAP** - нелинейный метод снижения размерности. Сохраняет как локальную, так и глобальную структуру. Быстрее t-SNE, есть `transform` для новых данных.

Идея: строит граф соседства в исходном пространстве, затем оптимизирует расположение точек в низкой размерности для сохранения структуры графа.

**Параметры:**
- `n_neighbors` - размер локального окружения (больше = больше глобальная структура)
- `min_dist` - минимальное расстояние между точками в проекции

### Токенизация (laba9)

<a id="токен"></a>
**Токен** - минимальная единица текста после разбиения. Может быть словом, символом или субсловом.

<a id="токенизация"></a>
**Токенизация** - разбиение текста на токены. Три основных подхода:

| Подход | Пример | Размер словаря | Токенов в тексте |
|---|---|---|---|
| **Word** | "I love NLP" → ["I", "love", "NLP"] | Большой | Мало |
| **Char** | "NLP" → ["N", "L", "P"] | Маленький | Много |
| **BPE** | "unbelievable" → ["un", "believ", "able"] | Средний | Средне |

<a id="word-tokenizer"></a>
**Word-level токенизация** - разбиение на слова и пунктуацию. `WordPunctTokenizer` из `nltk`. Перед использованием приводят к нижнему регистру (lower-case) чтобы "Love" и "love" были одним токеном.

<a id="char-tokenizer"></a>
**Char-level токенизация** - разбиение посимвольно. Маленький словарь, но тексты становятся очень длинными. Регистр обычно сохраняют.

<a id="bpe"></a>
**BPE (Byte-Pair Encoding)** - субсловная токенизация:
1. Начинает с посимвольного разбиения
2. Итеративно объединяет самые частые пары токенов
3. Останавливается при достижении нужного размера словаря

Компромисс между word и char: редкие слова разбиваются на части, частые остаются целыми.

<a id="стоп-слова"></a>
**Стоп-слова** - частые слова без смысловой нагрузки (the, a, is, and). Обычно удаляются перед анализом.

<a id="лемматизация"></a>
**Лемматизация** - приведение слова к словарной форме (running → run, better → good).

<a id="стемминг"></a>
**Стемминг** - грубое отсечение окончаний (running → run, runs → run). Быстрее лемматизации, менее точно.

### Признаки из текста

<a id="bag-of-words"></a>
**Bag of Words (BoW, мешок слов)** - для каждого текста создаётся вектор: сколько раз встречается каждое слово из словаря. Порядок слов и смысл игнорируются. Простой, но работает.

<a id="tf-idf"></a>
**TF-IDF** - взвешенный вариант BoW. Два компонента:
- **TF** (Term Frequency) - частота слова в данном тексте
- **IDF** (Inverse Document Frequency) - обратная частота слова во всех текстах

$$TF\text{-}IDF(w, d) = \frac{Count(w \in d)}{|d|} \cdot \log\frac{N}{Count(w \in D) + \alpha}$$

где $x$ - отдельный текст, $D$ - весь набор документов, $N$ - общее число документов, $\alpha$ - параметр сглаживания (обычно 1).

Редкие слова получают высокий вес (информативны), частые по всем текстам - низкий (стоп-слова). Лучше BoW для задач классификации.

### Эмбеддинги (laba10)

<a id="эмбеддинг"></a>
**Эмбеддинг (векторное представление)** - отображение слова (или текста) в числовой вектор. Близкие по смыслу слова - близкие векторы. Позволяет делать арифметику со словами: king - man + woman ≈ queen.

<a id="word2vec"></a>
**Word2Vec** - модель для обучения эмбеддингов слов на корпусе текстов. Обучается предсказывать слово по контексту (или контекст по слову). Слова с похожим окружением получают похожие векторы.

<a id="fasttext"></a>
**FastText** - расширение Word2Vec: использует субсловные (символьные) n-граммы. Умеет строить вектор для слова которого не было в обучении.

<a id="gensim"></a>
**gensim** - библиотека NLP для работы с эмбеддингами. Позволяет обучить Word2Vec/FastText на своём корпусе или загрузить предобученную модель.

<a id="косинусное-сходство"></a>
**Косинусное сходство** - мера близости двух векторов:

$$cos(\theta) = \frac{a \cdot b}{||a|| \cdot ||b||}$$

От -1 до 1. Используется для поиска похожих слов/документов в пространстве эмбеддингов.

<a id="roc-auc"></a>
**ROC-AUC** - метрика качества бинарного классификатора, устойчивая к дисбалансу классов.

Классификатор обычно выдаёт не класс, а **вероятность** (от 0 до 1). Порог решения (по умолчанию 0.5) определяет, что считать положительным. При разных порогах получаем разные значения TPR и FPR:

- **TPR (True Positive Rate = Recall)** - доля реально положительных, которых нашли: $TPR = \frac{TP}{TP+FN}$
- **FPR (False Positive Rate)** - доля реально отрицательных, которых ошибочно отнесли к положительным: $FPR = \frac{FP}{FP+TN}$

**ROC-кривая** - график TPR от FPR при всех возможных порогах от 0 до 1.

**AUC (Area Under Curve)** - площадь под ROC-кривой:
- 1.0 = идеальный классификатор (при некотором пороге TPR=1, FPR=0)
- 0.5 = случайное угадывание (диагональ на графике)
- < 0.5 = модель хуже случайного

Интерпретация: вероятность того, что случайно выбранный положительный пример получит более высокий score чем случайно выбранный отрицательный.

**Scatter map** - точечная карта: каждая точка = одно наблюдение с координатами. Используется для визуализации плотности и распределения событий.

<a id="geopandas-plot"></a>
**`gdf.plot()`** - статичная отрисовка GeoDataFrame. Параметры: `column` (столбец для цветовой кодировки), `legend`, `figsize`, `cmap` (цветовая схема).

### Интерактивная карта на plotly (go.Scattermap)

Для интерактивной карты используется `go.Scattermap` - точки на карте с hover и фильтрацией.

**Структура:**
```python
fig = make_subplots(
    specs=[[{"type": "map"}]]  # тип ячейки - карта
)

fig.add_trace(go.Scattermap(
    lat=df['pickup_latitude'],
    lon=df['pickup_longitude'],
    mode='markers',
    marker=dict(size=4, color='blue', opacity=0.6),
    customdata=custom_data,       # доп. данные для hover
    hovertemplate=hover_template  # шаблон подсказки
))

fig.update_layout(
    map=dict(
        style="carto-positron",           # стиль карты
        center=dict(lat=40.71, lon=-74.0), # центр
        zoom=10
    )
)
```

<a id="scattermap"></a>
**`go.Scattermap`** - точки на географической карте. Принимает `lat` и `lon`. Поддерживает `customdata` + `hovertemplate` для детального hover.

<a id="map-style"></a>
**`map.style`** - стиль подложки карты: `"carto-positron"` (светлая), `"open-street-map"`, `"carto-darkmatter"` (тёмная).

<a id="updatemenus-map"></a>
**`updatemenus`** - кнопки фильтрации. Каждая кнопка переключает видимость трейсов (`visible`). Позволяет смотреть данные по районам/группам без перерисовки.
